```python
# Import necessary libraries
import pandas as pd
from zipfile import ZipFile   # To extract files from the zip folder
from astropy.io import ascii  # To read the asc file
```

```python
# path of the zip folder
zip_file = '/content/SouthGermanCredit.zip'

# Extracting files
with ZipFile(zip_file, 'r') as zip:
  zip.printdir()

  print('Extracting files from zip folder...')
  zip.extractall()
```

```python
# reading the data from asc file
ascii_data = pd.read_table('/content/SouthGermanCredit.asc')
ascii_data
```

```python
# Converting the raw ascii data to a table
table = ascii.read('/content/SouthGermanCredit.asc')
table[:10]
```

```python
# Converting the table to a dataframe
df = pd.DataFrame(table.as_array())
df.head()
```

We see the heading of the columns are in German. We cange it into the correspnding English terms by refering the codetable.txt file.

```python
# Renaming columns
df.columns = ['status', 'duration', 'credit_history', 'purpose', 'amount', 'savings',
              'employment_duration', 'installment_rate', 'personal_status_sex', 'other_debtors', 
              'present_residence', 'property', 'age', 'other_installment_plans', 'housing', 'number_credits',
              'job', 'people_liable', 'telephone', 'foreign_worker', 'credit_risk']
```

```python
df.head()
```

```python
# Convert the dataframe to csv file
df.to_csv('German_credit.csv', index=None)
```
